```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# ステップ関数
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

```python
x = np.arange(-5, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# シグモイド関数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y 軸の範囲を指定
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
```
